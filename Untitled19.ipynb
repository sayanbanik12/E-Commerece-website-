```python
import pandas as pd
import numpy as np
import spacy
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("/content/fake_and_real_news.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df['label'].value_counts()
```

```python
df['label_num'] = df['label'].map({
    'Fake': 0,
    'Real': 1
})
```

```python
df.head()
```

```python
!python -m spacy download en_core_web_lg
```

```python
nlp = spacy.load("en_core_web_lg")
df['vector'] = df['Text'].apply(lambda text: nlp(text).vector)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df['vector'].values,
    df['label_num'],
    test_size = 0.2,
    random_state = 42
)
```

```python
X_train_2d = np.stack(X_train)
X_test_2d = np.stack(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB

scaler = MinMaxScaler()
scaled_train_embed = scaler.fit_transform(X_train_2d)
scaled_test_embed = scaler.transform(X_test_2d)

NB = MultinomialNB()
NB.fit(scaled_train_embed, y_train)
NBy_pred = NB.predict(scaled_test_embed)

print(classification_report(y_test, NBy_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

KN = KNeighborsClassifier(n_neighbors = 5, metric = 'euclidean')

KN.fit(X_train_2d, y_train)

KNy_pred = KN.predict(X_test_2d)

print(classification_report(y_test, KNy_pred))
```
